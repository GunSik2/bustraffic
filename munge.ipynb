```python
## 버스정류소 테이블 정리
import pandas as pd

state_bus = 'data/부천버스정류소현황.csv'
busstop_data = pd.read_csv(state_bus,  encoding='euc-kr')
busstop_data.shape
#busstop_data

df = pd.DataFrame(busstop_data, columns=['정류소명', '정류소id','WGS84위도','WGS84경도'])
busstop = df.rename(columns={"정류소명": "station", "정류소id": "stationid", 'WGS84위도': 'latitude','WGS84경도': 'longitude'})
busstop

busstop.to_csv('gold/busstop.csv', index=False)  
r = pd.read_csv('gold/busstop.csv')
r.shape
r
```

```python
usebus = 'data/out_alcard.txt'
usebus_data = pd.read_csv(usebus)
usebus_data.shape
usebus_data
```

```python
usebus = 'data/out_alcard.txt'
usebus_data = pd.read_csv(usebus)
usebus_data.shape
usebus_data


df = pd.DataFrame(usebus_data, columns=['ride_station_name','ride_time', 'gender'])
df

df = df.rename(columns={"ride_station_name": "station", "ride_time": "ride_time", 'gender': 'gender'})
df

ridebus = df[['ride_time', 'station', 'gender']]

ridebus.to_csv('gold/ridebus.csv', index=False)  
r = pd.read_csv('gold/ridebus.csv')
r
```

```python
usebus = 'data/out_cashbebee.txt'
usebus_data = pd.read_csv(usebus)
usebus_data.shape
usebus_data


df = pd.DataFrame(usebus_data, columns=['station_name','used_time'])
df

df = df.rename(columns={"station_name": "station", "used_time": "ride_time"})
ridebus = df[['ride_time', 'station']]
ridebus

ridebus.to_csv('gold/ridebus_cashbee.csv', index=False)  
r = pd.read_csv('gold/ridebus_cashbee.csv')
r
```

```python
busstop = pd.read_csv('gold/busstop.csv')
busstop

ridebus = pd.read_csv('gold/ridebus.csv')
ridebus

ridebusstop = pd.merge(left = ridebus , right = busstop, how = "inner", on = "station")
ridebusstop.to_csv('gold/ridebusstop.csv', index=False)  
r = pd.read_csv('gold/ridebusstop.csv')
r
```
